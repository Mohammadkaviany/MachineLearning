```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("/content/Real estate valuation data set.xlsx", usecols="C, H")
df
```

```python
x = df['X2 house age']
y = df["Y house price of unit area"]
```

```python
plt.plot(df['X2 house age'], df["Y house price of unit area"], 'o')
```

# single variable linear regression with SKLearn


```python
age = LinearRegression()
age.fit(df[['X2 house age']], df["Y house price of unit area"])
print(age.intercept_)
print(age.coef_)
```

```python
def line(theta0, theta1, x):
  y = np.array(theta0 + ( theta1 * x) )
  return y
```

```python
x = np.linspace(0, 50, 100)
y = 42.434697 + (-0.25148842 * x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df['X2 house age'], df["Y house price of unit area"], 'o', label= "Data")
plt.plot(x, line(age.intercept_, age.coef_, x), "-x", label = "Predicted Values")
plt.legend()
plt.grid()
```
