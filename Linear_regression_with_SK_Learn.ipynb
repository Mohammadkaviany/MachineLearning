```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("/content/Real estate valuation data set.xlsx", usecols="C, H")
df
```

```python
x = df['X2 house age']
y = df["Y house price of unit area"]
```

```python
plt.plot(df['X2 house age'], df["Y house price of unit area"], 'o')
```

# single variable linear regression with SKLearn


```python
age = LinearRegression()
age.fit(df[['X2 house age']], df["Y house price of unit area"])
print(age.intercept_)
print(age.coef_)
```

```python
def line(theta0, theta1, x):
  y = np.array(theta0 + ( theta1 * x) )
  return y
```

```python
x = np.linspace(0, 50, 100)
y = 42.434697 + (-0.25148842 * x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df['X2 house age'], df["Y house price of unit area"], 'o', label= "Data")
plt.plot(x, line(age.intercept_, age.coef_, x), "-x", label = "Predicted Values")
plt.legend()
plt.grid()
```

# اجرای مدل با در نظر گرفتن تمامی ویژگی ها

# Multiple Linear Regression With scikit-learn

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# بردار ایکس به عنوان بردار تمام ویژگی ها در نظر گرفته شده است
# همچنین بردار وای هم به عنوان برچسب درنظر گرفته شده است

```python
x = df.drop('Y house price of unit area',axis= 1)
y = df['Y house price of unit area']
```

# در نظر گرفتن 20 در از داده ها به عنوان داده های تست و نیز 80 د رصد به عنوان داده های ترین

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# یادگیری مدل از روی داده های ترین

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

# پیش بینی کردن مقادیر جدید و نیز بدست اوردن خط فیت شده و عرض از مبدا و شیب آن

```python
Predict = model.predict(X_test)
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df['X2 house age'], df["Y house price of unit area"], 'o', label= "Data")
plt.plot(x, line(model.intercept_, model.coef_, x), "-x", label = "Predicted Values")
plt.legend()
plt.grid()
```
